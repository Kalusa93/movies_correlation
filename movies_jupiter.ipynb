```python
# First let's import the packages we will use in this project
# You can do this all now or as you need them
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
import scipy
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (16,8)

pd.options.mode.chained_assignment = None



# Now we need to read in the data
df = pd.read_csv('movies.csv')
```

```python
df
```

```python
# We need to see if we have any missing data
# Let's loop through the data and see if there is anything missing

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Data Types for our columns

print(df.dtypes)
```

```python
# Are there any Outliers?

df.boxplot(column=['gross'])
```

```python
df.drop_duplicates()
```

```python
# Order our Data a little bit to see

df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
sns.regplot(x="budget", y="gross", data=df)
```

```python
sns.regplot(x="budget", y="gross", data=df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

```python
# We suspect a correlation between score and gross
```

```python
sns.regplot(x="score", y="gross", data=df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

```python
# Apparently it's not significant
```

```python
# Correlation Matrix between all numeric columns

df.corr(method ='pearson', numeric_only=True)
```

```python
# Another correlation method: Kendall
df.corr(method ='kendall', numeric_only=True)
```

```python
# Another correlation method: Spearman
df.corr(method ='spearman', numeric_only=True)
```

```python
# We'll keep working with the default method, Pearson
# For ease of view, we'll now use a correlation matrix with only numeric columns

correlation_matrix = df.corr(numeric_only = True)
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
```

```python
# There is a high correlation between budget/gross, and votes/gross
```

```python
# Looking at the top 15 companies by gross revenue

CompanyGrossSum = df.groupby('company')[["gross"]].sum()

CompanyGrossSumSorted = CompanyGrossSum.sort_values('gross', ascending = False)[:15]

CompanyGrossSumSorted = CompanyGrossSumSorted['gross'].astype('int64') 

CompanyGrossSumSorted
```

```python
df.groupby(['company', 'year'])[["gross"]].sum()
```

```python
# Top 15 most profitable companies by year

CompanyGrossSum = df.groupby(['company', 'year'])[["gross"]].sum()

CompanyGrossSumSorted = CompanyGrossSum.sort_values(['gross','company','year'], ascending = False)[:15]

CompanyGrossSumSorted = CompanyGrossSumSorted['gross'].astype('int64') 

CompanyGrossSumSorted
```

```python
# Scatter plot Budget vs Gross Earnings

plt.scatter(x=df['budget'], y=df['gross'], alpha=0.5)
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')
plt.show()
```

```python
df_numerized = df


for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
df_numerized.corr(method='pearson')
```

```python
correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Movies")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
# We confirm a high correlation between budget/gross, and votes/gross. And there are 
# no significant correlations between non numerical columns
```

```python
# Now we ask the question of which rating of movies is the most profitable. 
# We plot gross against the movies rating

sns.stripplot(x="rating", y="gross", data=df, hue="rating")
```

```python
# We can see that PG-13 movies are the most profitable, followed by PG ones.
```
